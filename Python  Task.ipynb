# **1. Print list in reverse order using a loop.**

```python
students = [{'ID':'10','Name':'Mia'},
 {'ID':'09','Name':'Aiden'},
  {'ID':'08','Name':'Zoe'},
   {'ID':'07','Name':'Ethan'},
    {'ID':'06','Name':'Ava'},
     {'ID':'05','Name':'Leo'},
      {'ID':'04','Name':'Grace'},
       {'ID':'03','Name':'Noah'},
        {'ID':'02','Name':'Lily'},
         {'ID':'01','Name':'Sam'}]
```

```python
for i in students[::-1]:
  print(i)
```

# **2. Use a loop to display elements from a given list present at an odd index position.**

```python
students = [{'ID':'10','Name':'Mia'},
 {'ID':'09','Name':'Aiden'},
  {'ID':'08','Name':'Zoe'},
   {'ID':'07','Name':'Ethan'},
    {'ID':'06','Name':'Ava'},
     {'ID':'05','Name':'Leo'},
      {'ID':'04','Name':'Grace'},
       {'ID':'03','Name':'Noah'},
        {'ID':'02','Name':'Lily'},
         {'ID':'01','Name':'Sam'}]
```

```python
for i in range(len(students)):
    if i % 2 != 0:
        print(students[i])
```
